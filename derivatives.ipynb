# Derivatives

We can use derivatives as a way to measure the instantaneous rate of change for a function. This allows us to find the marginal cost, marginal revenue, and marginal profit.

Let's say our company sells an app and let's say the average price of the app is $20 and we run all the back end in the cloud, so the cost is variable based on what we use (ignoring salary, building cost, etc for a simple example). In this example I will estimate the cost at 0.0001x^2, where x is the number of people that purchase the app.
We can write a function that defines revenue like this:

```python
def revenue(x):
    revenue = 20 * x - 0.0001 * x ** 2
    return revenue
```

Since we are going to want to know the incremental revenue of a user purchasing our app, we need to have a derivative function:

```python
def derivative(x, delta):
    deriv = (revenue(x + delta) - revenue(x)) / delta
    return deriv
```

Now that I have both of those functions, I want to plot them so that I can get a visual of how they behave.Â 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100000)
plt.plot(x, revenue(x))
plt.title('Revenue function: 20x - 0.0001x^2')
plt.show()
```

```python
plt.plot(x, derivative(x, 0.0001))
plt.title('Incremental Revenue')
plt.show()
```
